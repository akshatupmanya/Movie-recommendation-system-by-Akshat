```python
import ast
def conv(x):
  l=[]
  for i in ast.literal_eval(x):
    l.append(i['name'])
  return l
```

```python
mov['genres']= mov['genres'].apply(conv)
```

```python
mov['keywords']=mov['keywords'].apply(conv)
```

```python
def conv_cast(x):
  l=[]
  count=0
  for i in ast.literal_eval(x):
    if count!=2:
      l.append(i['name'])
      count+=1
    else:
      break
  return l
```

```python
mov['cast']=mov['cast'].apply(conv_cast)
```

```python
mov.head()
```

```python
def conv_crew(x):
  l=[]
  for i in ast.literal_eval(x):
    if i['job']== 'Director':
      l.append(i['name'])
  return l
```

```python
mov['crew']=mov['crew'].apply(conv_cast)
```
